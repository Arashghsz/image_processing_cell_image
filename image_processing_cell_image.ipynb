```python
import cv2 as cv 
import glob
import os
import numpy as np
from PIL import Image
from PIL import ImageFilter
from skimage.filters import roberts, sobel, scharr, prewitt, farid
from skimage import io
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

# Read Images from path

```python
list_images = []
folder_dir = '/content/drive/MyDrive/Multi-Media-p1/Cell-Images/'
for directory_path in glob.glob(folder_dir):
    for folder_dir in glob.glob(os.path.join(directory_path, "*.png")):
        img = cv.imread(folder_dir,0)       
        list_images.append(img)
```

```python
plt.imshow(list_images[0], cmap="gray")
```

```python
list_images[3].dtype
```

# Apply filter to all Images
I've imported 5 filters. just use one of them(roberts).

```python
# roberts_img = roberts(list_images[0])
filtered_images=[]
for i in list_images:
  filtered_images.append(roberts(i))
```

```python
plt.imshow(list_images[0], cmap='gray')
```

# Improving the contrast in all Images

```python
# equ = cv.equalizeHist(list_images[0])
equ_images=[]
for i in list_images:
  equ_images.append(cv.equalizeHist(i))
```

```python
cv2_imshow(equ_images[50])
cv.waitKey(0)
```

# Edge enhancing for one image(just test)

```python
# Open an already existing image
path = '/content/drive/MyDrive/Multi-Media-p1/Cell-Images/174793807517.png'
imageObject = Image.open(path)


edgeEnahnced = imageObject.filter(ImageFilter.EDGE_ENHANCE)
moreEdgeEnahnced = imageObject.filter(ImageFilter.EDGE_ENHANCE_MORE)
plt.imshow(imageObject)
plt.imshow(edgeEnahnced)
plt.imshow(moreEdgeEnahnced, cmap="gray")
plt.show()
```

# FINAL RESULTS 

```python
plt.figure(figsize=([20, 20]))
plt.subplot(141),plt.imshow(list_images[50], cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(142),plt.imshow(equ_images[50], cmap = 'gray')
plt.title('equalize hist'), plt.xticks([]), plt.yticks([])
plt.subplot(143),plt.imshow(filtered_images[50], cmap="gray")
plt.title('roberts filtering'), plt.xticks([]), plt.yticks([])
plt.subplot(144),plt.imshow(moreEdgeEnahnced)
plt.title('Result after edge enhancement'), plt.xticks([]), plt.yticks([])
plt.show()
```
